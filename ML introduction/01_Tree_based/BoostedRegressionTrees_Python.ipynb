```python
import xgboost as xgb
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

## Classification
Response: Species -> 3 classes

```python
brt = xgb.XGBClassifier(objective='multi:softprob', num_class=2, n_estimators = 50, use_label_encoder=False)
brt.fit(X, Y)
```

Make predictions:

```python
pred = brt.predict_proba(X)
```

Probabilities for each species:

```python
print(pred)
```

Variable importance:

```python
brt.feature_importances_
```

## Regression
Response: Sepal.Length -> continuous

```python
brt = xgb.XGBRegressor(objective='reg:squarederror', n_estimators = 50)
brt.fit(X[:,1:], X[:,0])
```

Make predictions:

```python
pred = brt.predict(X[:,1:])
print(pred)
```

Variable importance:

```python
brt.feature_importances_
```
