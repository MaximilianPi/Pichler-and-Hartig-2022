```python
from sklearn import datasets, svm
iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

## Classification
Response: Species -> 3 classes

```python
sv = svm.SVC(probability=True)
sv.fit(X, Y)
```

Make predictions:

```python
pred = sv.predict_proba(X)
```

Probabilities for each species:

```python
print(pred)
```

## Regression
Response: Sepal.Length -> continuous

```python
sv = svm.SVR()
sv.fit(X[:,1:], X[:,0])
```

Make predictions:

```python
pred = sv.predict(X[:,1:])
print(pred)
```
